```python
import pandas as pd
import yfinance as yf #yfinance library to access various financial data
from datetime import datetime
```

```python
start_date = datetime.now() - pd.DateOffset(months=3) #DateOffset to shift dates forward or backward
end_date = datetime.now()
```

```python
tickers = ['TSLA', 'TM', 'HMC','SPY'] #list of stock tickers
```

```python
df_list = [] #empty datalist to store multiple DataFrames in a list
```

```python

for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date)
    df_list.append(data) #add an element to a list.

df_list
```

```python
df = pd.concat(df_list, keys=tickers, names=['Ticker', 'Date']) #it will concatenate the DataFrames in df_list along a new index with levels 'Ticker' and 'Date', where 'Ticker' corresponds to the tickers from the tickers list, and 'Date' is the date index from the original DataFrames in df_list.
print(df.head())
df.shape
```

```python
df = df.reset_index()
df.head()
```

**Stock Market Performance (Last 3 Months)**

```python
import plotly.express as px
fig = px.line(df, x='Date',
              y='Close',
              color='Ticker',
              title="Stock Market Performance for the Last 3 Months")
fig.show()
```

**Stock Prices for differnt motor companies**

```python
fig = px.area(df, x='Date', y='Close', color='Ticker',
              facet_col='Ticker',
              labels={'Date':'Date', 'Close':'Closing Price', 'Ticker':'Company'},
              title='Stock Prices for SPY, TSLA, HMC, and TM')
fig.show()
```

**Moving Average: Stock Price Movement over a period of time**

```python
df['MA10'] = df.groupby('Ticker')['Close'].rolling(window=10).mean().reset_index(0, drop=True)
 #new column named 'MA10' to a DataFrame df. This new column contains the 10-day moving average of
 #the 'Close' prices for each unique 'Ticker' group
df['MA20'] = df.groupby('Ticker')['Close'].rolling(window=20).mean().reset_index(0, drop=True)

for ticker,group in df.groupby('Ticker'):
    print(f'Moving Averages for {ticker}')
    print(group[['MA10', 'MA20']]) #prints the columns 'MA10' and 'MA20' from the current group (group) DataFrame.
```

```python
for ticker, group in df.groupby('Ticker'):
    fig = px.line(group, x='Date', y=['Close', 'MA10', 'MA20'],
                  title=f"{ticker} Moving Averages")
    fig.show()
```

**Volatality in Stock Prices**

```python
df['Volatility'] = df.groupby('Ticker')['Close'].pct_change().rolling(window=10).std().reset_index(0, drop=True)
# 10-day rolling standard deviation of daily returns for each stock.
fig = px.line(df, x='Date', y='Volatility',
              color='Ticker',
             title='Volatility of All Companies')
print(df)
fig.show()
```

**Correlation between Stock Prices**

```python
# create a DataFrame with the stock prices of SPY and TSLA
SPY = df.loc[df['Ticker'] == 'SPY', ['Date', 'Close']].rename(columns={'Close': 'SPY'})
TSLA = df.loc[df['Ticker'] == 'TSLA', ['Date', 'Close']].rename(columns={'Close': 'TSLA'})

df_corr = pd.merge(SPY,TSLA, on='Date')

# create a scatter plot to visualize the correlation

fig = px.scatter(df_corr, x='SPY', y='TSLA',
                 trendline='ols',  #ordinary least squares
                 title='Correlation between SPY and TSLA',
                 color= 'TSLA',
                 color_continuous_scale='Turbo')


fig.show()

# create a DataFrame with the stock prices of SPY and TM
SPY = df.loc[df['Ticker'] == 'SPY', ['Date', 'Close']].rename(columns={'Close': 'SPY'})
TM = df.loc[df['Ticker'] == 'TM', ['Date', 'Close']].rename(columns={'Close': 'TM'})
df_corr = pd.merge(SPY,TM, on='Date')

# create a scatter plot to visualize the correlation

fig = px.scatter(df_corr, x='SPY', y='TM',
                 trendline='ols',  #ordinary least squares
                 title='Correlation between SPY and TM',
                  color= 'TM',
                 color_continuous_scale='Turbo')

fig.show()

# create a DataFrame with the stock prices of SPY and HMC
SPY = df.loc[df['Ticker'] == 'SPY', ['Date', 'Close']].rename(columns={'Close': 'SPY'})
HMC = df.loc[df['Ticker'] == 'HMC', ['Date', 'Close']].rename(columns={'Close': 'HMC'})
df_corr = pd.merge(SPY,HMC, on='Date')

# create a scatter plot to visualize the correlation

fig = px.scatter(df_corr, x='SPY', y='HMC',
                 trendline='ols',  #ordinary least squares
                 title='Correlation between SPY and HMC',
                  color= 'HMC',
                 color_continuous_scale='Turbo')

fig.show()
```
